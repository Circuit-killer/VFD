# Variable Frequency Drive math

Calculating signal characteristics and energy losses for Variable Frequency Drive (VFD). Topology: Pulse Width Modulated (PWM), Voltage/Frequncy (V/f) controll

```python
%pylab inline
```

```python
def phase(x, F, V, P):
    "Return sinewave function of given properties"
    return (V*sin((x+radians(P))*F))

def sawTooth(x, F, A):
    "returns sawtoof function"
    return (A*(x*F-floor(x*F)))
def genPWM(sine, saw):
    "generates pwm"
    dif=sine-saw
    out=[]
    for i in range(len(dif)):
        if (dif[i]) > 0:
            a=1
        else:
            a=0
        out.append(a)
    
    return asarray(out)
```

```python
freq = 50
voltage = 230*sqrt(2)
PWMFreq= 1000
resoliution= 100.0*PWMFreq #will cause terible aliasing if less than 2x PWMFreq
x = linspace(0, 2*pi/freq, resoliution)
```

```python
plot(x, (phase(x, freq, voltage, 0)))
plot(x, (phase(x, freq, voltage, 120)))
plot(x, (phase(x, freq, voltage, -120)))
plot(x, (sawTooth(x, PWMFreq, 2*voltage)-voltage))
```

pwm output for all 3 phases

```python
subplot(3, 1, 1)
plot(x, (genPWM(phase(x, freq, 1, 0), sawTooth(x, PWMFreq, 2)-1)))
plot(x, (phase(x, freq, 1, 0)))
ylim(-1.2, 1.2)
subplot(3, 1, 2)
plot(x, (genPWM(phase(x, freq, 1, 120), sawTooth(x, PWMFreq, 2)-1)))
plot(x, (phase(x, freq, 1, 120)))
ylim(-1.2, 1.2)
subplot(3, 1, 3)
plot(x, (genPWM(phase(x, freq, 1, -120), sawTooth(x, PWMFreq, 2)-1)))
plot(x, (phase(x, freq, 1, -120)))
ylim(-1.2, 1.2)
```

```python
pwm1=genPWM(phase(x, freq, 1, 0), sawTooth(x, PWMFreq, 2)-1)
dutyCycle=count_nonzero(pwm1)/resoliution
print 'Duty Cycle: ', dutyCycle
edges= (pwm1[1:] - pwm1[:-1])
rel= 800e-6 #rising edge loss
fel= 500e-6 #falling edge loss
switchingLoss= (2*rel*edges[(edges>0)].size + 2*fel*edges[(edges<0)].size)*freq #J/s or W for 1 phase
print 'SwitchingLoss: ', switchingLoss, 'W/ph'
Amps=5 #current per phase (A)
ffwdCurrent=2.5  #forword Voltage per transistor (V)
print 'Powerloss due to current: ', ffwdCurrent*Amps*2*dutyCycle, 'W/ph'
heat= (dutyCycle*Amps*ffwdCurrent*2 +switchingLoss) *3
print 'Total heat generated: ', heat, 'W'
print 'Conditions. motor freqency:', freq, 'Hz. PWM frequncy: ', PWMFreq, 'Hz., Current: ', Amps, 'A/phase'
#as shown in this example duty cycle is always 50% for each phase, and switching losses increase drasticaly with PWM freaquency 
```
